```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

The momentum equation in vector form of a velocity field $\vec{v}$ is:

$$
\dfrac{\partial \vec{v}}{\partial t} + (\vec{v} \cdot \nabla)\vec{v} = - \dfrac{1}{\rho}\nabla p +\nu \nabla^2 \vec{v}
$$

This equation in vectorial form has three equations: one for each velocity componenets $(u, w, v)$ - solving in 2D we will get rid of one of them. Poisson's equation will be solved to complete the full set of 3 equations that will be solved:

$$
\dfrac{\partial u}{\partial t} + u \dfrac{\partial u}{\partial x} + v \dfrac{\partial u}{\partial y} = - \dfrac{1}{\rho} \dfrac{\partial p}{\partial x} + \nu \left( \dfrac{\partial^2 u}{\partial x^2} + \dfrac{\partial^2 u}{\partial y^2} \right) \\
$$

$$
\dfrac{\partial v}{\partial t} + u \dfrac{\partial v}{\partial x} + v \dfrac{\partial v}{\partial y} = - \dfrac{1}{\rho} \dfrac{\partial p}{\partial y} + \nu \left( \dfrac{\partial^2 v}{\partial x^2} + \dfrac{\partial^2 v}{\partial y^2} \right) \\
$$

$$
\nabla^2 p^{n+1} = \rho \dfrac{\nabla \cdot \mathbf{u}^n}{\Delta t} - \rho \nabla \cdot (\mathbf{u}^n \cdot \nabla \mathbf u^n) + \mu\nabla^2(\nabla\cdot \mathbf{u}^n) \xrightarrow[]{\text{w/ } \mu \text{ small}}  \dfrac{\partial^2 p}{\partial x^2} + \dfrac{\partial^2 p}{\partial y^2} = \rho \left( \dfrac{1}{\Delta t} \left( \dfrac{\partial u}{\partial x} + \dfrac{\partial v}{\partial y} \right) \right) - \rho \left( \dfrac{\partial u}{\partial x}\dfrac{\partial u}{\partial x}+2\dfrac{\partial u}{\partial y}\dfrac{\partial v}{\partial x}+\dfrac{\partial v}{\partial y }\dfrac{\partial v}{\partial y} \right) 
$$

The equations must be discretized as done in previous steps, having:

$$
\frac{u_{i,j}^{n+1}-u_{i,j}^{n}}{\Delta t}+u_{i,j}^{n}\frac{u_{i,j}^{n}-u_{i-1,j}^{n}}{\Delta x}+v_{i,j}^{n}\frac{u_{i,j}^{n}-u_{i,j-1}^{n}}{\Delta y} =-\frac{1}{\rho}\frac{p_{i+1,j}^{n}-p_{i-1,j}^{n}}{2\Delta x}+\nu\left(\frac{u_{i+1,j}^{n}-2u_{i,j}^{n}+u_{i-1,j}^{n}}{\Delta x^2}+\frac{u_{i,j+1}^{n}-2u_{i,j}^{n}+u_{i,j-1}^{n}}{\Delta y^2}\right)
$$

$$
\frac{v_{i,j}^{n+1}-v_{i,j}^{n}}{\Delta t}+u_{i,j}^{n}\frac{v_{i,j}^{n}-v_{i-1,j}^{n}}{\Delta x}+v_{i,j}^{n}\frac{v_{i,j}^{n}-v_{i,j-1}^{n}}{\Delta y} =-\frac{1}{\rho}\frac{p_{i,j+1}^{n}-p_{i,j-1}^{n}}{2\Delta y}
+\nu\left(\frac{v_{i+1,j}^{n}-2v_{i,j}^{n}+v_{i-1,j}^{n}}{\Delta x^2}+\frac{v_{i,j+1}^{n}-2v_{i,j}^{n}+v_{i,j-1}^{n}}{\Delta y^2}\right)
$$

$$
\frac{p_{i+1,j}^{n}-2p_{i,j}^{n}+p_{i-1,j}^{n}}{\Delta x^2}+\frac{p_{i,j+1}^{n}-2p_{i,j}^{n}+p_{i,j-1}^{n}}{\Delta y^2} 
=\rho\left[\frac{1}{\Delta t}\left(\frac{u_{i+1,j}-u_{i-1,j}}{2\Delta x}+\frac{v_{i,j+1}-v_{i,j-1}}{2\Delta y}\right)-\frac{u_{i+1,j}-u_{i-1,j}}{2\Delta x}\frac{u_{i+1,j}-u_{i-1,j}}{2\Delta x}\right. -2\frac{u_{i,j+1}-u_{i,j-1}}{2\Delta y}\frac{v_{i+1,j}-v_{i-1,j}}{2\Delta x}-\left.\frac{v_{i,j+1}-v_{i,j-1}}{2\Delta y}\frac{v_{i,j+1}-v_{i,j-1}}{2\Delta y}\right]
$$

Solving now for each one of the variables:

$$
u_{i,j}^{n+1} = u_{i,j}^{n} - u_{i,j}^{n}\frac{\Delta t}{\Delta x}(u_{i,j}^{n}-u_{i-1,j}^{n}) - v_{i,j}^{n}\frac{\Delta t}{\Delta y}(u_{i,j}^{n}-u_{i,j-1}^{n}) -\frac{\Delta t}{\rho}\frac{p_{i+1,j}^{n}-p_{i-1,j}^{n}}{2\Delta x}+\nu\Delta t\left(\frac{u_{i+1,j}^{n}-2u_{i,j}^{n}+u_{i-1,j}^{n}}{\Delta x^2}+\frac{u_{i,j+1}^{n}-2u_{i,j}^{n}+u_{i,j-1}^{n}}{\Delta y^2}\right)
$$


$$
v_{i,j}^{n+1} = v_{i,j}^{n}-u_{i,j}^{n}\frac{\Delta t}{\Delta x}(v_{i,j}^{n}-v_{i-1,j}^{n}) - v_{i,j}^{n}\frac{\Delta t}{\Delta y}(v_{i,j}^{n}-v_{i,j-1}^{n}) -\frac{\Delta t}{\rho}\frac{p_{i,j+1}^{n}-p_{i,j-1}^{n}}{2\Delta y}
+\nu\Delta t\left(\frac{v_{i+1,j}^{n}-2v_{i,j}^{n}+v_{i-1,j}^{n}}{\Delta x^2}+\frac{v_{i,j+1}^{n}-2v_{i,j}^{n}+v_{i,j-1}^{n}}{\Delta y^2}\right)
$$


$$
p_{i,j}^{n}=\frac{(p_{i+1,j}^{n}+p_{i-1,j}^{n})\Delta y^2+(p_{i,j+1}^{n}+p_{i,j-1}^{n})\Delta x^2}{2(\Delta x^2+\Delta y^2)}-\frac{\Delta x^2\Delta y^2}{2(\Delta x^2+\Delta y^2)} \left[\frac{\rho}{\Delta t}\left(\frac{u_{i+1,j}-u_{i-1,j}}{2\Delta x}+\frac{v_{i,j+1}-v_{i,j-1}}{2\Delta y}\right)-\frac{u_{i+1,j}-u_{i-1,j}}{2\Delta x}\frac{u_{i+1,j}-u_{i-1,j}}{2\Delta x}\right. -2\frac{u_{i,j+1}-u_{i,j-1}}{2\Delta y}\frac{v_{i+1,j}-v_{i-1,j}}{2\Delta x}-\left.\frac{v_{i,j+1}-v_{i,j-1}}{2\Delta y}\frac{v_{i,j+1}-v_{i,j-1}}{2\Delta y}\right]
$$

The conditions applied to the discretized equations are:
* **Initial conditions**:
$$
u, v, p = 0 \text{ everywhere}
$$


* **Velocity boundary conditions**:
$$
\begin{Bmatrix}
    u = 1  & at & y = 2 \\
    u, v = 0 & at & \text{all the other boundaries}
\end{Bmatrix}
$$


* **Pressure boundary conditions**:
$$
\begin{Bmatrix}
    \dfrac{\partial p}{\partial y} = 0 & at & y = 0 \\
    p = 0 & at & y = 2 \\
    \dfrac{\partial p}{\partial x} = 0 & at & x = 0,\ 2 \\
\end{Bmatrix}
$$

```python
nx = 41
ny = 41
nt = 500
nit = 50
c = 1
dx = 2 / (nx - 1)
dy = 2 / (ny - 1)

x = np.linspace(0, 2, nx)
y = np.linspace(0, 2, ny)

X, Y = np.meshgrid(x, y)

rho = 1
nu = 0.1
dt = 0.001

u = np.zeros((ny, nx))
v = np.zeros((ny, nx))
p = np.zeros((ny, nx))
b = np.zeros((ny, nx))
```

```python
def build_up_b(b, rho, dt, u, v, dx, dy):
    
    b[1:-1, 1:-1] = (rho * (1 / dt * ((u[1:-1, 2:] - u[1:-1, 0:-2]) / (2 * dx) + (v[2:, 1:-1] - v[0:-2, 1:-1]) / (2 * dy)) -
                    ((u[1:-1, 2:] - u[1:-1, 0:-2]) / (2 * dx))**2 -2 * ((u[2:, 1:-1] - u[0:-2, 1:-1]) / (2 * dy) *
                    (v[1:-1, 2:] - v[1:-1, 0:-2]) / (2 * dx)) - ((v[2:, 1:-1] - v[0:-2, 1:-1]) / (2 * dy))**2))

    return b
```

```python
def pressure_poisson(p, dx, dy, b):
    pn = np.empty_like(p)
    pn = p.copy()
    
    for q in range(nit):
        pn = p.copy()
        p[1:-1, 1:-1] = (((pn[1:-1, 2:] + pn[1:-1, 0:-2]) * dy**2 + (pn[2:, 1:-1] + pn[0:-2, 1:-1]) * dx**2) /
                          (2 * (dx**2 + dy**2)) - dx**2 * dy**2 / (2 * (dx**2 + dy**2)) * b[1:-1,1:-1])

        p[:, -1] = p[:, -2] ##dp/dy = 0 at x = 2
        p[0, :] = p[1, :]  ##dp/dy = 0 at y = 0
        p[:, 0] = p[:, 1]    ##dp/dx = 0 at x = 0
        p[-1, :] = 0        ##p = 0 at y = 2
        
    return p
```

```python
def cavity_flow(nt, u, v, dt, dx, dy, p, rho, nu):
    un = np.empty_like(u)
    vn = np.empty_like(v)
    b = np.zeros((ny, nx))
    
    for n in range(nt):
        un = u.copy()
        vn = v.copy()
        
        b = build_up_b(b, rho, dt, u, v, dx, dy)
        p = pressure_poisson(p, dx, dy, b)
        
        u[1:-1, 1:-1] = (un[1:-1, 1:-1]- un[1:-1, 1:-1] * dt / dx * (un[1:-1, 1:-1] - un[1:-1, 0:-2]) - vn[1:-1, 1:-1] * dt / dy * (un[1:-1, 1:-1] - un[0:-2, 1:-1]) -
                         dt / (2 * rho * dx) * (p[1:-1, 2:] - p[1:-1, 0:-2]) + nu * (dt / dx**2 * (un[1:-1, 2:] - 2 * un[1:-1, 1:-1] + un[1:-1, 0:-2]) + dt / dy**2 *
                        (un[2:, 1:-1] - 2 * un[1:-1, 1:-1] + un[0:-2, 1:-1])))

        v[1:-1,1:-1] = (vn[1:-1, 1:-1] - un[1:-1, 1:-1] * dt / dx * (vn[1:-1, 1:-1] - vn[1:-1, 0:-2]) - vn[1:-1, 1:-1] * dt / dy * (vn[1:-1, 1:-1] - vn[0:-2, 1:-1]) -
                        dt / (2 * rho * dy) * (p[2:, 1:-1] - p[0:-2, 1:-1]) + nu * (dt / dx**2 * (vn[1:-1, 2:] - 2 * vn[1:-1, 1:-1] + vn[1:-1, 0:-2]) + dt / dy**2 * 
                        vn[2:, 1:-1] - 2 * vn[1:-1, 1:-1] + vn[0:-2, 1:-1]))

        u[0, :] = 0
        u[:, 0] = 0
        u[:, -1] = 0
        u[-1, :] = 1    #set velocity on cavity lid equal to 1
        
        v[0, :] = 0
        v[-1, :]=0
        v[:, 0] = 0
        v[:, -1] = 0
        
        
    return u, v, p
```

```python
u = np.zeros((ny, nx))
v = np.zeros((ny, nx))
p = np.zeros((ny, nx))
b = np.zeros((ny, nx))
nt = 1000
u, v, p = cavity_flow(nt, u, v, dt, dx, dy, p, rho, nu)
```

```python
fig = plt.figure(figsize=(8.5,7), dpi=100)

# plotting the pressure field as a contour
plt.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)  
plt.colorbar()

# plotting velocity field
plt.quiver(X[::2, ::2], Y[::2, ::2], u[::2, ::2], v[::2, ::2]) 

plt.xlabel('X')
plt.ylabel('Y')
```

## Grid refinement

```python
grid = np.empty_like(p)
```

```python
fig = plt.figure(figsize=(5,5), dpi=100)
plt.scatter(X, Y, color='black', s=2)  
```

```python
[gradX, gradY] = np.gradient(p)

fig = plt.figure(figsize=(8.7,4), dpi=100)
ax1 = fig.add_subplot(1, 2, 1)
ax1.axis('equal')
ax1.contourf(X, Y, gradX, alpha=0.5, cmap=cm.jet)  
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_title('Gradient in X')

ax2 = fig.add_subplot(1, 2, 2)
ax2.axis('equal')
ax2.contourf(X, Y, gradY, alpha=0.5, cmap=cm.jet)  
ax2.set_xlabel('$x$')
ax2.set_title('Gradient in X')
```

```python
gradT = np.sqrt(gradX**2+gradY**2)/np.max(np.sqrt(gradX**2+gradY**2))

fig = plt.figure(figsize=(5, 4), dpi=100)
plt.contourf(X, Y, gradT, 20, alpha=0.5, cmap=cm.jet)  
plt.colorbar()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Average gradient')
plt.scatter(X, Y, color = 'black', s=2)  
```

```python
filterVal = 0.08

bigGrad = [gradT > filterVal]

fig = plt.figure(figsize=(4,4), dpi=100)
plt.scatter(X, Y, bigGrad[0], color = 'black') 
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Big gradient nodes')
```

```python
newX = np.linspace(0, 2, 2*nx-1)
newY = np.linspace(0, 2, 2*ny-1)

nX, nY = np.meshgrid(newX, newY)

fig = plt.figure(figsize=(8.7,4), dpi=100)
ax1 = fig.add_subplot(1, 2, 1)
ax1.axis('equal')
ax1.scatter(X, Y, color='black', s=1)  
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_title('Previous mesh')

ax2 = fig.add_subplot(1, 2, 2)
ax2.axis('equal')
ax2.scatter(nX, nY, color='black', s=1)  
ax2.set_xlabel('$x$')

ax2.set_title('New mesh')
```

```python
mask = np.zeros((2*nx-1,2*nx-1))

for i in range(np.shape(mask)[0]):
    for j in range(np.shape(mask)[1]):
        if i%2 == 0:
            if j%2 == 0:
                mask[i,j] = 1
```

```python
fig = plt.figure(figsize=(8.7,4), dpi=100)
ax1 = fig.add_subplot(1, 2, 1)
ax1.axis('equal')
ax1.scatter(X, Y, color='black', s=1)
ax1.scatter(X, Y, bigGrad, color='red')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_title('Previous mesh')

ax2 = fig.add_subplot(1, 2, 2)
ax2.axis('equal')
ax2.scatter(nX, nY, mask, color='black')  
ax2.set_xlabel('$x$')

ax2.set_title('New mesh')
```

```python
for i in range(np.shape(bigGrad)[1]):
    for j in range(np.shape(bigGrad)[2]):
        if bigGrad[0][i,j] == True:
            if j == 0:
                mask[2*i, 2*j] = 1
                mask[2*i-1, 2*j] = 1                
            elif i == 0:
                mask[2*i, 2*j] = 1
                mask[2*i, 2*j-1] = 1                
            else:
                mask[2*i, 2*j] = 1
                mask[2*i-1, 2*j] = 1
                mask[2*i, 2*j-1] = 1
                mask[2*i-1, 2*j-1] = 1
```

```python
fig = plt.figure(figsize=(8.7,4), dpi=100)
ax1 = fig.add_subplot(1, 2, 1)
ax1.axis('equal')
ax1.scatter(X, Y, color='black', s=1)
ax1.scatter(X, Y, bigGrad, color='red')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_title('Previous mesh')

ax2 = fig.add_subplot(1, 2, 2)
ax2.axis('equal')
ax2.scatter(nX, nY, mask, color='black')  
ax2.set_xlabel('$x$')

ax2.set_title('New mesh')
```
