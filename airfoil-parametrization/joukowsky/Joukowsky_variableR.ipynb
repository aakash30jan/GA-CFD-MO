```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def joukowsky(x_cent,y_cent,radius):
    # definition of the circle parameters
    center = np.array([x_cent,y_cent])
    radius1 = radius

    # calculation of the circle coordinates 
    angle = np.linspace(0,2*np.pi,720)
    chi1 = center[0] + radius1*np.cos(angle)
    eta1 = center[1] + radius1*np.sin(angle)

    # calculations of the Joukowsky transform
    x1 = ((chi1)*(chi1**2+eta1**2+1))/(chi1**2+eta1**2)
    y1 = ((eta1)*(chi1**2+eta1**2-1))/(chi1**2+eta1**2)

    # figure declaration
    plt.figure(figsize=(12,6))
    
    # zeta-plane 
    ax = plt.subplot(1, 2, 1)
    ax.plot(chi1,eta1,'g',label='Circle')
    ax.plot(center[0],center[1],'kx', markersize=10,label='Center')
    ax.plot([-1,1],[0,0],'rh',markersize=10,label='Reference Points')
    ax.axis('equal')
    ax.set_xlim([-2,2])
    ax.set_ylim([-2,2])
    ax.grid(True)
    ax.set_xlabel(r"$\chi$",size=20)
    ax.set_ylabel(r"$\eta$",size=20)
    ax.set_title(r'$\zeta$-plane', fontsize=22)
    leg = ax.legend(loc='lower left', fontsize=14)
    leg.get_frame().set_alpha(1)
    ax.tick_params(axis = 'both', labelsize = 20)

    # z-plane 
    ax = plt.subplot(1, 2, 2)
    ax.axis('equal')
    ax.plot(x1,y1,'g',label='Transform1')
    ax.set_xlim([4,-4])
    ax.set_ylim([-4,4])
    ax.set_xlabel(r"$x$",size=20)
    ax.set_ylabel(r"$y$",size=20)
    ax.set_title(r'$z$-plane', fontsize=22)
    ax.tick_params(axis = 'both', labelsize = 20)

    # save the figure as .png
    plt.savefig('variableR.png', bbox_inches='tight')
```

```python
joukowsky(0.5,0.1,1.0)
```
